```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
import sys, getopt
```

```python
data = requests.get("https://fbref.com/en/comps/9/stats/Premier-League-Stats")
soup = BeautifulSoup(data.text)
teams_table = soup.select('table.stats_table')[0]
teams_table
```

```python
links = teams_table.find_all('a')
links = [link.get('href') for link in links]
links = [link for link in links if '/squads/' in link]
```

```python
team_links = [f"https://fbref.com{team}" for team in links]
team_links
```

```python
team_link = team_links[0]
data = requests.get(team_link)
players = pd.read_html(data.text, match="Standard Stats")
players[0]
```
